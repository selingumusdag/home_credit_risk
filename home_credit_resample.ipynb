```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport

application_train=pd.read_csv('./home_credit_data/application_train.csv')
application_train.info()
```

```python
application_train.head()
```

```python
application_train['TARGET'].value_counts()
```

```python
from sklearn.utils import resample

target_zero = application_train[application_train.TARGET == 0]
target_one = application_train[application_train.TARGET == 1]

train_resample_one = resample(target_one,
                                     replace = True,
                                     n_samples = 5000,
                                     random_state = 111)
train_resample_zero = resample(target_zero,
                                     replace = True,
                                     n_samples = 5000,
                                     random_state = 111)

app_train = pd.concat([train_resample_one, train_resample_zero])
app_train.TARGET.value_counts()
```

```python
app_train.info()
```

```python
profile=ProfileReport(app_train, title='Application Train')
```

```python
profile
```

```python
app_train.to_csv('./home_credit_data/app_train.csv')
```
